```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'Earthquakes_database.csv'
```

```python
df = pd.read_csv(file_path)
```

```python
# task 3
```

```python
pivot_table = df.pivot_table(values='Magnitude', index='Year', aggfunc='mean')
```

```python
melted_table = pivot_table.reset_index().melt(id_vars=['Year'], value_name='Average Magnitude')
```

```python
print("Pivot Table: Average Magnitude of Earthquakes Per Year")
print(pivot_table.head())
print("\nMelted Table: Long Format")
melted_table.head()
```

```python
# task 4
```

```python
# 1 Line Diagram: Number of earthquakes over time
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

df['Year'] = df['Date'].dt.year
earthquakes_by_year = df.groupby('Year').size()

plt.figure(figsize=(10, 6))
earthquakes_by_year.plot(kind='line')
plt.title('Number of Earthquakes Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Earthquakes')
plt.grid(True)
plt.show()
```

```python
# 2 Bar Chart: Number of earthquakes by magnitude type
magnitude_counts = df['Magnitude Type'].value_counts()

plt.figure(figsize=(10, 6))
magnitude_counts.plot(kind='bar')
plt.title('Number of Earthquakes by Magnitude Type')
plt.xlabel('Magnitude Type')
plt.ylabel('Number of Earthquakes')
plt.xticks(rotation=45)
plt.show()
```

```python
# 3 Circular Diagram (Pie Chart): Distribution of earthquake types
type_counts = df['Type'].value_counts()

plt.figure(figsize=(8, 8))
type_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Earthquake Types')
plt.ylabel('')
plt.show()
```

```python
# 4 Histogram: Distribution of earthquake depths
# Convert 'Depth' to numeric, coercing errors to NaN, then replace NaNs with 0
df['Depth'] = pd.to_numeric(df['Depth'], errors='coerce').fillna(0)

plt.figure(figsize=(10, 6))
sns.histplot(df['Depth'], bins=30, kde=True)
plt.title('Distribution of Earthquake Depths')
plt.xlabel('Depth (km)')
plt.ylabel('Frequency')
plt.show()
```
